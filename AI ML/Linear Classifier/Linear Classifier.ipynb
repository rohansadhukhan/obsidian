```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()

x = iris.data
y = iris.target
```

```python
_, ax = plt.subplots()

scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target) # ploting based on sepal length and width

ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])

_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
_, ax = plt.subplots()

scatter = ax.scatter(x[:, 0], x[:, 1], c=y) # ploting based on sepal length and width

ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])

_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
# we are first learn with 2 classes and 2 feature
x = x[0:100,0:2]
y = y[0:100]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)
```

```python
scaler = StandardScaler()

scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
model = SGDClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
training_accuracy = model.score(x_train, y_train)

print(f'Prediction: \t{y_pred}')
print(f'Actual: \t{y_test}')
print(f'Accuracy: \t{accuracy}')
print(f'Training Accuracy: \t{training_accuracy}')
```
